```python
tau = 1+I

#################################################
#Here are all the points and matrices:

Id = matrix([[1,0,0],[0,1,0],[0,0,1]])

Pinf = matrix([[1],[0],[0]])


P0 = matrix([[0],[0],[1]])
I0 = matrix([[0,0,1],[0,-1,0],[1,0,0]])

P2 = matrix([[(-1+I)/2],[1],[1]])
A2 = matrix([[-1,-2,2+I],[1+I,2+I,-1-I],[1+I,1+I,-I]])

P41 = matrix([[I/2],[0],[1]])
A41 = matrix([[I,0,1],[0,-1,0],[2,0,-I]])

P42 = matrix([[-1/2],[1],[1]])
A42 = matrix([[-1,-2,2],[2,3,-2],[2,2,-1]])

#Gamma_infty
Tv = matrix([[1,0,I],[0,1,0],[0,0,1]])
T2 = matrix([[1,-2,-2],[0,1,2],[0,0,1]])
Ttau = matrix([[1,-1+I,-1],[0,1,1+I],[0,0,1]])
R = matrix([[1,0,0],[0,I,0],[0,0,1]])


matrices_gamma_infinity = [['R',R],['R^-1',R^-1],['T2',T2],['T2^-1',T2^-1],['Ttau',Ttau],['Ttau^-1',Ttau^-1],['Tv',Tv],['Tv^-1',Tv^-1]]


matrices = [['I0',I0],['A2',A2],['A41',A41],['A42',A42]]
points = [['P0',P0],['P2',P2],['P41',P41],['P42',P42]]
```

```python
# these are the scalars who clear the denominators of the prism points of each of their respective depths.  Technically d1 should contain 1, not empty, but that would be silly and would just mean we mutiply by 1 a lot.
d1 = []
d2 = [tau]
d4 = [2]
depths = [d2,d4]
```

```python
#Compute the inner product of two vectors
def ip(v1,v2):
    return v2[2].conjugate()*v1[0] + v2[1].conjugate()*v1[1] + v2[0].conjugate()*v1[2]

#Test if a (rational) number (not a vector) is in the ring of integers.
def IsIntegral(n):
    if (n.real() in ZZ) & (n.imag() in ZZ):
        return True
    else:
        return False

#Given a vector v, find its primitive integral lift.  Only works on points of depth up to 4 because what it does is multiply by all possible scalars that could clear the denominator and then checks if the coordinates are integral.
def PrimIntLift(v):
    if IsIntegral(v[0][0]) & IsIntegral(v[1][0]) & IsIntegral(v[2][0]):
        return v
    for d in depths:
        for q in d:
            w = v*q
            if IsIntegral(w[0][0]) & IsIntegral(w[1][0]) & IsIntegral(w[2][0]):
                return w
    print("depth greater than 4")
    return False
   
#Find the level between two points.  Only works on points of depth up to 4.
def level(v1,v2):
    return ip(PrimIntLift(v1),PrimIntLift(v2)).norm()

#Find the depth of a point.  Only works on points of depth up to 4.
def depth(v):
    PIL = PrimIntLift(v)
    if PIL == False:
        return False
    return PIL[2][0].norm()


#Computes the Cygan distance between two points.  What form are these points in?  Clearly not vector form.... 
def Cygan(p1,p2):
    return ((p1[0]-p2[0]).norm()^2 + abs(p1[1]-p2[1]+2*ImPart(p2[0].conjugate()*p1[0]))^2)^(1/4)
    
def vectCygan(v1,v2):
    nv1 = v1*v1[2][0]^(-1)
    nv2 = v2*v2[2][0]^(-1)
    prod = 2*ip(nv1,nv2)
    return (prod.norm())^(1/4)
```

```python
#given a point p representing an orbit, the radius within which the others centers can lie is sqrt(2)/depth(p)^(1/4) + sqrt(2).  It will, of course be less if the other center is not level 1.
r1 = 2*sqrt(2.)
r2 = sqrt(2.)/2^(1/4.) + sqrt(2.)
r4 = sqrt(2.)/4^(1/4.) + sqrt(2.)
```

```python
#Get enough stuff to cover the whole base

def vertDisp(m,n):
    p = T2^m*Ttau^n*P0
    return p[0][0].imag()

def stringlet(g,p):
    if p == 0:
        return 'Id'
    if p == 1:
        return g
    return g+'^'+str(p)

def startn(m):
    if m == -3:
        return 1
    if m == -2:
        return -1
    if m == -1:
        return -2
    if m == 0:
        return -3
    if m == 1:
        return -4
    if m == 2:
        return -5
    if m == 3:
        return -6
        
def endn(m):
    if m == -3:
        return 5
    if m == -2:
        return 5
    if m == -1:
        return 4
    if m == 0:
        return 3
    if m == 1:
        return 2
    if m == 2:
        return 1
    if m == 3:
        return -1

Colors = []
for i in range(100):
    Colors.append([])

m = -3
while m <= 3:
    n = startn(m)
    while n <= endn(m):
        d = vertDisp(m,n)
        Colors[d].append([stringlet('T2',m)+'*'+stringlet('Ttau',n), T2^m*Ttau^n])
        n +=1
    m+=1
```

```python
#now make everyone go vertically

VerticalColors = []
for i in range(100):
    VerticalColors.append([])

for i in range(50):
    pos = Colors[i]
    neg = Colors[-i]
    for p in pos:
        n = -9 - i
        while n <= 8 - i:
                g = [stringlet('Tv',n)+'*'+p[0],Tv^n*p[1]]
                VerticalColors[i].append(g)
                n += 1
        for m in neg:
            n = -9 + i
            while (n <= 8 + i) and not(i == 0):
                g = [stringlet('Tv',n)+'*'+m[0],Tv^n*m[1]]
                VerticalColors[-i].append(g)
                n += 1
```

```python
#Finally, multiply by all powers of R. ON THE RIGHT!

RotatedVerticalColors = []
for i in range(100):
    RotatedVerticalColors.append([])

for vc in VerticalColors:
    i = VerticalColors.index(vc)
    for g in vc:
        RotatedVerticalColors[i].append(g)
        j=1
        while j <= 3:
            RotatedVerticalColors[i].append([g[0]+'*'+stringlet('R',j),g[1]*R^j])
            j += 1
```

```python
#Now, put 'em all together.
TooMany = []
for rvc in RotatedVerticalColors:
    for g in rvc:
        TooMany.append(g)
        
len(TooMany)
```

```python
cycles = []
for g in TooMany:
    #print TooMany.index(g)
    for P in points:
        #print 'P=points['+str(points.index(P))+']'
        for Q in points:
            #print 'Q=points['+str(points.index(Q))+']'
            gQ = g[1]*Q[1]
            l = level(P[1],gQ)
            #print 'level='+str(l)
            if l <= 4:
                aa = points.index(P)
                c = points.index(Q)
                gam1Pb = matrices[aa][1]^(-1)*g[1]*matrices[c][1]*Pinf
                d = depth(gam1Pb)
                if d <= 4:
                    C = [ matrices[aa], matrices[c], g]
                    cycles.append(C)

len(cycles)
```

```python
#Spidercrawl around, outward from the identity, in a nice pattern, attempting to identify which gamma_infinity translate we have on our hands.  It will be super inefficient.

def spidercrawl(C):
    #C is an element of cycles, consisting of A_a, A_c, and gamma_3
    #returns the pair A_b, gamma_1
    
    found = 0
    gamm1Pb = C[0][1]^(-1)*C[2][1]*C[1][1]*Pinf
    s = gamm1Pb[2][0]
    if not(s == 0):
        gamm1Pb = gamm1Pb*s^(-1)
    else:
        found = 1
        return [['Id',Id],['Id',Id]]
    d = depth(gamm1Pb)
    levdpoints = []
    levdmats = []
    if d == 1:
        levdpoints = [points[0]]
        levdmats = [matrices[0]]
    if d == 2:
        levdpoints = [points[1]]
        levdmats = [matrices[1]]
    if d == 4:
        levdpoints = [points[2],points[3]]
        levdmats = [matrices[2],matrices[3]]
    if len(levdpoints) == 0:
        print 'A_c^-1*gamma_3^-1*A_a*Pinf has depth greater than 4'
        return []
    
    for p in levdpoints:
        if gamm1Pb == p[1]:
            found = 1
            return [ levdmats[levdpoints.index(p)], ['Id',Id]]

    W = matrices_gamma_infinity
    #wordlength = 1
    while found == 0:
        next = []
        for w in W:
            for p in levdpoints:
                candidate = w[1]*p[1]
                s = candidate[2][0]
                if not(s == 1):
                    candidate = candidate*s^(-1)
                if candidate == gamm1Pb:
                    found = 1
                    return [levdmats[levdpoints.index(p)],w]
            for m in matrices_gamma_infinity:
                next.append([m[0]+'*'+w[0],m[1]*w[1] ])
        #if wordlength == 7:
         #   print 'words are getting long, do this one by hand.'
          #  return []
        W = next
```

```python
def findRelation(C):
    
    B = spidercrawl(C)
    #print cycles.index(C)
    if B == []:
        print 'cycles['+str(cycles.index(C))+']'
        return
    gamma = C[1][1]^-1*C[2][1]^-1*C[0][1]*B[1][1]*B[0][1]
    found = 0
    
    if gamma == Id:
        found = 1
        return '('+C[1][0]+')^-1*('+C[2][0]+')^-1*('+C[0][0]+')*('+B[1][0]+')*('+B[0][0]+') == Id'
    if gamma == I*Id:
        found = 1
        return '('+C[1][0]+')^-1*('+C[2][0]+')^-1*('+C[0][0]+')*('+B[1][0]+')*('+B[0][0]+') == I*Id'
    if gamma == -Id:
        found = 1
        return '('+C[1][0]+')^-1*('+C[2][0]+')^-1*('+C[0][0]+')*('+B[1][0]+')*('+B[0][0]+') == -Id'
    if gamma == -I*Id:
        found = 1
        return '('+C[1][0]+')^-1*('+C[2][0]+')^-1*('+C[0][0]+')*('+B[1][0]+')*('+B[0][0]+') == -I*Id'
    
    W = matrices_gamma_infinity
    while found == 0:
        next = []
        for w in W:
            if gamma == w[1]:
                found = 1
                return '('+C[1][0]+')^-1*('+C[2][0]+')^-1*('+C[0][0]+')*('+B[1][0]+')*('+B[0][0]+') == '+w[0]
            if gamma == I*w[1]:
                found = 1
                return '('+C[1][0]+')^-1*('+C[2][0]+')^-1*('+C[0][0]+')*('+B[1][0]+')*('+B[0][0]+') == I*'+w[0]
            if gamma == -w[1]:
                found = 1
                return '('+C[1][0]+')^-1*('+C[2][0]+')^-1*('+C[0][0]+')*('+B[1][0]+')*('+B[0][0]+') == -'+w[0]
            if gamma == -I*w[1]:
                found = 1
                return '('+C[1][0]+')^-1*('+C[2][0]+')^-1*('+C[0][0]+')*('+B[1][0]+')*('+B[0][0]+') == -I*'+w[0]
            for m in matrices_gamma_infinity:
                next.append([m[0]+'*'+w[0],m[1]*w[1] ])
        W = next
```
